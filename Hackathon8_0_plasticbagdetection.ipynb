```python
from roboflow import Roboflow
import supervision as sv
import cv2

rf = Roboflow(api_key="lSPfTZYyrRFbc2ewO7Xl")
project = rf.workspace().project("octa-pda4c")
model = project.version(1).model
img_path = ("/content/drive/MyDrive/Hackathon8.0/trash_all.jpg")
result = model.predict(img_path, confidence=40).json()

labels = [item["class"] for item in result["predictions"]]

detections = sv.Detections.from_roboflow(result)

label_annotator = sv.LabelAnnotator()
mask_annotator = sv.MaskAnnotator()

image = cv2.imread(img_path)

annotated_image = mask_annotator.annotate(
    scene=image, detections=detections)
annotated_image = label_annotator.annotate(
    scene=annotated_image, detections=detections, labels=labels)

sv.plot_image(image=annotated_image, size=(12, 12))
```

```python
from skimage import io
from roboflow import Roboflow
import supervision as sv
import cv2

# Load and display the initial image
image_path = "/content/drive/MyDrive/Hackathon8.0/bottleandbag.jpg"
image = io.imread(image_path)
io.imshow(image)

# Initialize Roboflow and Supervision
rf = Roboflow(api_key="lSPfTZYyrRFbc2ewO7Xl")
project = rf.workspace().project("underwater_debris")
model = project.version(1).model

# Predict
result = model.predict(image_path, confidence=40, overlap=30).json()

# Filter predictions for class 'pbag' only
filtered_predictions = [item for item in result["predictions"] if item["class"] == "pbag"]

# Proceed only if there are filtered predictions
if filtered_predictions:
    # Update result with filtered predictions
    result["predictions"] = filtered_predictions

    # Generate detections from filtered predictions
    detections = sv.Detections.from_roboflow(result)

    # Dynamically generate labels list from filtered predictions
    labels = [item["class"] for item in filtered_predictions]

    # Load the image using cv2 (for annotation)
    image = cv2.imread(image_path)

    # Annotation
    label_annotator = sv.LabelAnnotator()
    bounding_box_annotator = sv.BoxAnnotator()

    # Annotate image with bounding boxes and labels for 'pbag' detections
    annotated_image = bounding_box_annotator.annotate(scene=image, detections=detections)
    annotated_image = label_annotator.annotate(scene=annotated_image, detections=detections, labels=labels)

    # Display the annotated image
    sv.plot_image(image=annotated_image, size=(6, 6))
else:
    print("No 'pbag' detections found.")
```
